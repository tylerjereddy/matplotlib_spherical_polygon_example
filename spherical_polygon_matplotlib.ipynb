```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import cPickle as pickle
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
#data for 30 spherical polygons on a unit sphere
list_spherical_polygon_coords = pickle.load(open('thirty_spherical_polygon_arrays.p', 'rb'))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
for spherical_polygon in list_spherical_polygon_coords:
    random_color = np.random.rand(3)
    polygon = Poly3DCollection([spherical_polygon], alpha = 1.0)
    polygon.set_color(random_color)
    ax.add_collection(polygon)
    ax.scatter(spherical_polygon[...,0], spherical_polygon[...,1], spherical_polygon[...,2], c = 'k') 
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
ax.set_zlim(-1,1)
fig.set_size_inches(8,8)
```
